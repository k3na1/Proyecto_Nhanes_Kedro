```python
# Limpiar Estado_Entrevista (ya no la necesitamos como variable predictora, cumplió su rol de filtro)
df_demo = df_demo.drop(columns=['Estado_Entrevista'])
```

```python
# Limpiar Nivel_Educacion: Códigos 7 (Refused) y 9 (Don't know) → NaN
df_demo['Nivel_Educacion'] = df_demo['Nivel_Educacion'].replace([7.0, 9.0], np.nan)
# Imputar con la moda por grupo de edad y género
df_demo['age_group_temp'] = pd.cut(df_demo['Edad'], bins=[17, 35, 50, 65, 120])
df_demo['Nivel_Educacion'] = df_demo.groupby(['Genero', 'age_group_temp'])['Nivel_Educacion'].transform(
    lambda x: x.fillna(x.mode()[0] if not x.mode().empty else np.nan)
)
```

```python
# Limpiar Estado_Civil: Códigos 77 (Refused) y 99 (Don't know) → NaN
df_demo['Estado_Civil'] = df_demo['Estado_Civil'].replace([77.0, 99.0], np.nan)
# Imputar con la moda por grupo de edad y género
df_demo['Estado_Civil'] = df_demo.groupby(['Genero', 'age_group_temp'])['Estado_Civil'].transform(
    lambda x: x.fillna(x.mode()[0] if not x.mode().empty else np.nan)
)
```

```python
# Limpiar Ratio_Ingresos_Familiares: Nulos directos (sin código especial en NHANES)
# Imputar con la mediana por grupo de edad y etnia (más representativo económicamente)
df_demo['Ratio_Ingresos_Familiares'] = df_demo.groupby(['Etnia', 'age_group_temp'])['Ratio_Ingresos_Familiares'].transform(
    lambda x: x.fillna(x.median())
)
```

```python
# Eliminar columna temporal de grupos de edad
df_demo = df_demo.drop(columns=['age_group_temp'])
```

### Limpieza df_alq

```python
df_alq.info()
```

```python
df_alq.head()
```

```python
# Transformamos los valores raros en la columna Frecuencia_Alcohol

# Reemplazar el placeholder microscópico de SAS (~5.4e-79) por NaN reales
SAS_NAN = 5.397605346934028e-79
for col in ['Frecuencia_Alcohol', 'Promedio_Bebidas_Alcohol']:
    # Usamos np.isclose para detectar números casi idénticos al placeholder
    mask = np.isclose(df_alq[col], SAS_NAN, rtol=1e-10, atol=1e-10)
    df_alq.loc[mask, col] = np.nan
```

```python
# Reemplazar códigos de error tradicionales de NHANES (77, 99, 777, 999) por NaN
df_alq['Frecuencia_Alcohol'] = df_alq['Frecuencia_Alcohol'].replace([77.0, 99.0], np.nan)
df_alq['Promedio_Bebidas_Alcohol'] = df_alq['Promedio_Bebidas_Alcohol'].replace([777.0, 999.0], np.nan)
```

```python
# LÓGICA CRUZADA: Si Tomo_Alcohol == 2.0 (Nunca bebió), entonces frecuencia y cantidad son 0
# Nota: Frecuencia_Alcohol usa el código 0 (o a veces 10, depende del ciclo) para "Nunca en el último año". 
# Usaremos 0 como nuestra base matemática para "sin consumo".
df_alq.loc[df_alq['Tomo_Alcohol'] == 2.0, 'Frecuencia_Alcohol'] = 0
df_alq.loc[df_alq['Tomo_Alcohol'] == 2.0, 'Promedio_Bebidas_Alcohol'] = 0
# Además, si alguien dice que su frecuencia en el último año es "Nunca" (código 0 en Frecuencia_Alcohol),
# su promedio de bebidas diarias ese año (Promedio_Bebidas_Alcohol) debe ser 0.
df_alq.loc[df_alq['Frecuencia_Alcohol'] == 0, 'Promedio_Bebidas_Alcohol'] = 0
print("\nNulos DESPUÉS de usar Tomo_Alcohol:")
print(df_alq.isnull().sum())
```

1. **Estandarización de Nulos:** Se convirtieron los códigos de "Se niega" o "No sabe" (7, 9, 77, 99...) a verdaderos `NaN`.
2. **Lógica Cruzada:** Si un paciente reportó que "Nunca ha bebido" (`Tomo_Alcohol == 2`), se asumió matemáticamente un consumo nulo, asignando `0` a su frecuencia (`Frecuencia_Alcohol`) y a su cantidad diaria (`Promedio_Bebidas_Alcohol`). Esto recuperó aproximadamente 1,000 datos que figuraban como faltantes estructurales.
3. **Imputación Estratificada (Pendiente):** Los verdaderos nulos restantes se imputarán agrupando a los pacientes por su género y edad.

```python
# Traer datos demográficos temporalmente
demo_temp = catalog.load("p_demo_primary")[['SEQN', 'Genero', 'Edad']]
df_alq = df_alq.merge(demo_temp, on='SEQN', how='left')

# Crear grupos de edad (ej. 18-35, 35-50, 50-65, 65+) para que la imputación sea más precisa
df_alq['age_group'] = pd.cut(df_alq['Edad'], bins=[17, 35, 50, 65, 120])

# Funciones de imputación
def impute_mode(x):
    # Rellena con el valor más repetido (moda). Si todo es NaN, usa la mediana.
    return x.fillna(x.mode()[0] if not x.mode().empty else np.nan)

def impute_median(x):
    # Rellena con el valor central del grupo
    return x.fillna(x.median())

# Imputar Tomo_Alcohol (Alguna vez bebió) con la moda de su grupo de género y edad
df_alq['Tomo_Alcohol'] = df_alq.groupby(['Genero', 'age_group'])['Tomo_Alcohol'].transform(impute_mode)

# Imputar Frecuencia_Alcohol y Promedio_Bebidas_Alcohol con la mediana de su grupo
df_alq['Frecuencia_Alcohol'] = df_alq.groupby(['Genero', 'age_group'])['Frecuencia_Alcohol'].transform(impute_median)
df_alq['Promedio_Bebidas_Alcohol'] = df_alq.groupby(['Genero', 'age_group'])['Promedio_Bebidas_Alcohol'].transform(impute_median)

# IMPORTANTE: Si después de imputar Tomo_Alcohol, un nulo resultó ser "No bebe" (2.0),
# nos aseguramos de que su frecuencia y cantidad sean 0.
df_alq.loc[df_alq['Tomo_Alcohol'] == 2.0, 'Frecuencia_Alcohol'] = 0
df_alq.loc[df_alq['Tomo_Alcohol'] == 2.0, 'Promedio_Bebidas_Alcohol'] = 0

# Eliminar las columnas demográficas temporales
df_alq = df_alq.drop(columns=['Genero', 'Edad', 'age_group'])

print("Nulos finales en P_ALQ:")
print(df_alq.isnull().sum())
```

```python
# Limpiar códigos de rechazo/no sabe en todas las variables
for col in ['Diagnostico_Diabetes', 'Prediabetes', 'Usa_Insulina', 'Tiene_Retinopatia']:
    df_diq[col] = df_diq[col].replace([7.0, 9.0], np.nan)
    
df_diq['Edad_Diagnostico_Diabetes'] = df_diq['Edad_Diagnostico_Diabetes'].replace([777.0, 999.0], np.nan)
df_diq['Edad_Diagnostico_Diabetes'] = df_diq['Edad_Diagnostico_Diabetes'].replace([666.0], 0) # 666 significa diagnosticado con menos de 1 año
```

```python
# Aplicamos Ingenieria de caracteristicas y logica cruzada

# Estado_Diabetes (0=Normal, 1=Prediabetes, 2=Diabetes)
df_diq['Estado_Diabetes'] = 0 # Base: Normal
df_diq.loc[(df_diq['Diagnostico_Diabetes'] == 3.0) | (df_diq['Prediabetes'] == 1.0), 'Estado_Diabetes'] = 1
df_diq.loc[df_diq['Diagnostico_Diabetes'] == 1.0, 'Estado_Diabetes'] = 2
```

```python
# Anios_Diabetes (Si no es diabético = 0. Si lo es: Edad Actual - Edad Diagnóstico)
df_diq['Anios_Diabetes'] = 0
mask_diabetico = (df_diq['Estado_Diabetes'] == 2)
df_diq.loc[mask_diabetico, 'Anios_Diabetes'] = df_diq['Edad'] - df_diq['Edad_Diagnostico_Diabetes']
```

```python
# Usa_Insulina (1=Sí, 0=No). Asumimos 0 si no es diabético.
df_diq['Usa_Insulina'] = 0
df_diq.loc[mask_diabetico & (df_diq['Usa_Insulina'] == 1.0), 'Usa_Insulina'] = 1
```

```python
# Tiene_Retinopatia (1=Sí, 0=No). Asumimos 0 si no es diabético.
df_diq['Tiene_Retinopatia'] = 0
df_diq.loc[mask_diabetico & (df_diq['Tiene_Retinopatia'] == 1.0), 'Tiene_Retinopatia'] = 1
```

```python
# Eliminar las columnas originales (ya extrajimos su valor) y la edad temporal
df_diq = df_diq.drop(columns=['Diagnostico_Diabetes', 'Prediabetes', 'Edad_Diagnostico_Diabetes', 'Usa_Insulina', 'Tiene_Retinopatia', 'Edad'])
```

```python
# Tratamiento de nulos restantes (Ej Diabéticos que no saben su edad de diagnóstico)
# Imputaremos Anios_Diabetes con la mediana de los que SÍ son diabéticos
mediana_anios = df_diq.loc[df_diq['Anios_Diabetes'] > 0, 'Anios_Diabetes'].median()
df_diq['Anios_Diabetes'] = df_diq['Anios_Diabetes'].fillna(mediana_anios)
# Si Anios_Diabetes resulta negativo por algún error de tipeo en la encuesta, lo fijamos en 0
df_diq.loc[df_diq['Anios_Diabetes'] < 0, 'Anios_Diabetes'] = 0
print("Vista previa del nuevo dataset limpio:")
display(df_diq.head())
print("\nNulos finales en df_diq:")
print(df_diq.isnull().sum())
```

   - Se unificó el diagnóstico en la variable `Estado_Diabetes` (0=Normal, 1=Prediabetes, 2=Diabetes).
   - Se calculó `Anios_Diabetes` restando la edad de diagnóstico a la edad actual. A los no diabéticos se les asignó `0`.
   - Se crearon variables binarias (1/0) para `Usa_Insulina` y `Tiene_Retinopatia`, asumiendo lógicamente `0` para personas sanas.
2. **Estandarización y Limpieza:** Se eliminaron las respuestas "No sabe" / "Se negó". Los valores faltantes en edad de diagnóstico (para los verdaderos diabéticos) fueron imputados con la mediana de años de enfermedad del resto de diabéticos.

```python
# Limpiar códigos de rechazo/no sabe en todas las columnas
columnas_mcq = [col for col in df_mcq.columns if col != 'SEQN']
for col in columnas_mcq:
    # Convertimos a 1 si el diagnóstico es afirmativo, de lo contrario 0
    df_mcq[col] = df_mcq[col].apply(lambda x: 1 if x == 1.0 else 0)
# Mapear y renombrar a variables descriptivas
df_mcq_limpio = pd.DataFrame()
df_mcq_limpio['SEQN'] = df_mcq['SEQN']
df_mcq_limpio['Tiene_Infarto'] = df_mcq['Tiene_Infarto']
df_mcq_limpio['Tiene_Derrame'] = df_mcq['Tiene_Derrame']
df_mcq_limpio['Tiene_Insuficiencia_Cardiaca'] = df_mcq['Insuficiencia_Cardiaca']
# Agrupamos Enfermedad Coronaria y Angina ya que fisiológicamente son casi lo mismo
df_mcq_limpio['Tiene_Enfermedad_Coronaria'] = df_mcq[['Enfermedad_Coronaria', 'Angina_Pecho']].max(axis=1)
df_mcq_limpio['Tiene_Cancer'] = df_mcq['Tiene_Cancer']
df_mcq_limpio['Tiene_EPOC'] = df_mcq['Tiene_EPOC']
df_mcq_limpio['Tiene_Enfermedad_Hepatica'] = df_mcq['Enfermedad_Hepatica']
# Verificar
df_mcq = df_mcq_limpio
print("Vista previa del nuevo dataset detallado:")
display(df_mcq.head())
print("\nNulos finales en df_mcq:")
print(df_mcq.isnull().sum())
```

1. **Transformación:** Las respuestas afirmativas se mapearon a `1`, y cualquier otra respuesta (no, no sabe, se negó) a `0`.
2. **Mapeo Detallado:** 
  - Se aislaron eventos críticos como `Tiene_Infarto` y `Tiene_Derrame`.

```python
# Limpiar códigos técnicos
# Fumo_100_Cigarrillos: 1=Sí, 2=No, 7=Refused, 9=Don't know
df_smq['Fumo_100_Cigarrillos'] = df_smq['Fumo_100_Cigarrillos'].replace([7.0, 9.0], np.nan)
# Fuma_Actualmente: 1=Todos los días, 2=Algunos días, 3=En absoluto, 7=Refused, 9=Don't know
df_smq['Fuma_Actualmente'] = df_smq['Fuma_Actualmente'].replace([7.0, 9.0], np.nan)
# Cigarrillos_Diarios_Ex_Fumador: cigarrillos/día.
# Edad_Inicio_Fumar: edad de inicio. 77/99 = Refused/Don't know
df_smq['Edad_Inicio_Fumar'] = df_smq['Edad_Inicio_Fumar'].replace([77.0, 99.0], np.nan)
```

```python
# INGENIERÍA DE CARACTERÍSTICAS (Manejo del Skip Pattern)
# Estado_Fumador (0=Nunca fumó, 1=Ex-fumador, 2=Fumador activo)
df_smq['Estado_Fumador'] = 0  # Base: Nunca fumó (incluye implícitamente a los Fumo_100_Cigarrillos == 2)
df_smq.loc[df_smq['Fuma_Actualmente'] == 3.0, 'Estado_Fumador'] = 1  # Dejó de fumar
df_smq.loc[df_smq['Fuma_Actualmente'].isin([1.0, 2.0]), 'Estado_Fumador'] = 2  # Fuma activamente
```

```python
# Cigarrillos_Diarios
# (Combinando Ex-fumadores y Activos)
df_smq['Cigarrillos_Diarios'] = 0 # Base: 0 para no fumadores
df_smq.loc[df_smq['Estado_Fumador'] == 1, 'Cigarrillos_Diarios'] = df_smq['Cigarrillos_Diarios_Ex_Fumador'] # Ex-fumadores
df_smq.loc[df_smq['Estado_Fumador'] == 2, 'Cigarrillos_Diarios'] = df_smq['Cigarrillos_Diarios_Activo'] # Activos

# Si algún fumador (activo o ex) no dijo cuántos fumaba, imputamos con la mediana
mediana_cigarrillos = df_smq.loc[df_smq['Cigarrillos_Diarios'] > 0, 'Cigarrillos_Diarios'].median()
mask_fumo = df_smq['Estado_Fumador'].isin([1, 2])
df_smq.loc[mask_fumo & df_smq['Cigarrillos_Diarios'].isna(), 'Cigarrillos_Diarios'] = mediana_cigarrillos
```

```python
# Anios_Fumando
df_smq['Anios_Fumando'] = 0
mask_fumo = df_smq['Estado_Fumador'].isin([1, 2])
df_smq.loc[mask_fumo, 'Anios_Fumando'] = df_smq['Edad'] - df_smq['Edad_Inicio_Fumar']
```

```python
# Imputación para los pocos fumadores que no dijeron su edad de inicio
mediana_anios_fumo = df_smq.loc[df_smq['Anios_Fumando'] > 0, 'Anios_Fumando'].median()
df_smq.loc[mask_fumo & df_smq['Anios_Fumando'].isna(), 'Anios_Fumando'] = mediana_anios_fumo
```

```python
# Corregir valores negativos (errores de tipeo donde edad de inicio > edad actual)
df_smq.loc[df_smq['Anios_Fumando'] < 0, 'Anios_Fumando'] = 0
```

```python
# Conservar solo las variables finales
df_smq = df_smq[['SEQN', 'Estado_Fumador', 'Cigarrillos_Diarios', 'Anios_Fumando']]
print("\nNulos finales en df_smq:")
print(df_smq.isnull().sum())
display(df_smq.head())
```

1. **Manejo del Skip Pattern (`Fumo_100_Cigarrillos`):** Se mapearon los códigos técnicos (No sabe/Rehusó) a nulos. Los `NaN` presentes de forma natural en las otras columnas corresponden al flujo legítimo del cuestionario, donde se omiten preguntas a las personas que nunca han fumado al menos 100 cigarrillos.
2. **Ingeniería de Características:**
   - **`Estado_Fumador`**: Se creó una escala ordinal (0=Nunca, 1=Ex-fumador, 2=Activo) basada en si fumaban actualmente (`Fuma_Actualmente`). Aquellos que no pasaron la pregunta filtro asumen el valor 0.
   - **`Cigarrillos_Diarios`**: Se consolidaron dos variables distintas del CDC. Para los ex-fumadores se usó `Cigarrillos_Diarios_Ex_Fumador` (cuántos fumaba cuando lo dejó) y para los activos `Cigarrillos_Diarios_Activo` (promedio de los últimos 30 días). Se forzó a `0` para los no fumadores.
   - **`Anios_Fumando`**: Se calculó restando la Edad de Inicio (`Edad_Inicio_Fumar`) a la Edad Actual. Se forzó a `0` para no fumadores y se corrigieron posibles edades negativas producto de errores de tipeo en el cuestionario original.
3. **Imputación Residual:** Los escasos fumadores activos o pasivos que tenían `NaN` en su dosis diaria o en su edad de inicio fueron imputados de forma segura usando la **mediana** calculada exclusivamente a partir del resto de los fumadores.
4. **Resultado Final:** Dataset robusto con **0 nulos**, combinando 5 variables de encuesta complejas en 3 variables sintéticas directas y limpias.

```python
from functools import reduce
# 1. Lista de todos los dataframes de variables independientes (X)
dataframes_X = [
    df_demo, df_alq, df_bmx, df_bpxo, df_diq, df_mcq, df_paq, df_slq, df_smq,
    df_ghb, df_hdl, df_tchol, df_trigly, df_biopro
]

# Unimos todas las X primero usando SEQN (Left Join)
df_master_x = reduce(lambda left, right: pd.merge(left, right, on='SEQN', how='left'), dataframes_X)
```

```python
# 2. Imputación de Triglicéridos/LDL (Para los que no ayunaron)
for col in ['Trigliceridos', 'Colesterol_LDL']:
    mediana = df_master_x[col].median()
    df_master_x[col] = df_master_x[col].fillna(mediana)
```

```python
# 3. Llamado al Catálogo de Kedro para traer la tabla de Mortalidad (nuestra 'y')
df_mort = catalog.load("p_mort_primary")
# Merge con el dataset maestro
df_final = pd.merge(df_master_x, df_mort, on='SEQN', how='left')
```

```python
# 4. REGLA DE NEGOCIO: Quedarnos SOLAMENTE con los que tienen registro de mortalidad
df_final = df_final.dropna(subset=['Estado_Mortalidad'])
# Imprimir Resultados
print(f"Dimensiones previas (Universo NHANES total): {df_master_x.shape}")
print(f"Dimensiones FINALES (Solo pacientes con seguimiento de mortalidad): {df_final.shape}")
print("\nDistribución de la variable objetivo (Estado_Mortalidad):")
print(df_final['Estado_Mortalidad'].value_counts())
# Comprobación de nulos absolutos en la tabla final
print(f"\nNulos totales restantes en el Master Dataset: {df_final.isnull().sum().sum()}")
display(df_final.head())
```
